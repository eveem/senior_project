```python
import pandas as pd
from pythainlp.tokenize import word_tokenize
```

```python
df = pd.read_csv('./final_process/law_text.csv')
df.head()
```

```python
# import deepcut

wc = []
ts = []
tns = []

for idx, row in df.iterrows():
#     words = deepcut.tokenize(row['Text'])
    words = word_tokenize(row['Text'])
    wc.append(len(words))
    ts.append(' '.join(words))
    temp = ''
    for i in words:
        if i != ' ':
            temp += i
    tns.append(temp)
```

```python
df['word_count'] = wc
df['text_no_space'] = tns
df['text_space_pythai'] = ts
```

```python
filtered_dict = {}

for idx, row in word_freq_df.iterrows():
    if len(row[0]) > 1:
        if row[0] in filtered_dict:
            filtered_dict[row[0]] += row[1]
        else:
            filtered_dict[row[0]] = row[1]
```

```python
count_freq = {}

for text in tns:
    for w in filtered_dict:
        x = text.count(w)
        if w in count_freq:
            count_freq[w] += x
        else:
            count_freq[w] = x
```

```python
import operator

sorted_all_words = sorted(count_freq.items(), key=operator.itemgetter(1))
## remove zero
sorted_all_words = sorted_all_words[8:]
```

```python
sorted_dict = {}

for i in sorted_all_words:
    sorted_dict[i[0]] = i[1]
```

```python
del sorted_dict['ไปผู้ใด']
del sorted_dict['นั้นเพื่อ']
```

```python
count_by_NO = {}

for idx, row in df.iterrows():
    in_this_NO = set()
    
    for w in sorted_dict:
        if w in row['Text']:
            in_this_NO.add(w)

    for w in in_this_NO:
        if w not in count_by_NO:
            count_by_NO[w] = 1
        else:
            count_by_NO[w] += 1
```

```python
sorted_by_law_NO = sorted(count_by_NO.items(), key=operator.itemgetter(1))
```

```python
sorted_by_law = {}

for i in sorted_by_law_NO:
    sorted_by_law[i[0]] = i[1]
```

```python
## 10 percent cut off

not_use10 = []

for w in sorted_by_law:
    x = sorted_by_law[w]/130 * 100
    if x > 10:
        not_use10.append(w)
```

```python
## 20 percent cut off

not_use20 = []

for w in sorted_by_law:
    x = sorted_by_law[w]/130 * 100
    if x > 20:
        not_use20.append(w)
```

```python
## 30 percent cut off

not_use30 = []

for w in sorted_by_law:
    x = sorted_by_law[w]/130 * 100
    if x > 30:
        not_use30.append(w)
```

```python
## 40 percent cut off

not_use40 = []

for w in sorted_by_law:
    x = sorted_by_law[w]/130 * 100
    if x > 40:
        not_use40.append(w)
```

```python
## 50 percent cut off

not_use50 = []

for w in sorted_by_law:
    x = sorted_by_law[w]/130 * 100
    if x > 50:
        not_use50.append(w)
```

```python
## use sorted_dict

def generate_words_matrix(not_use_arr):
    words_for_matrix = []
    for i in sorted_dict:
        if i not in not_use_arr:
            words_for_matrix.append(i)
    return words_for_matrix
```

```python
## change not use percent here

use_this_dict_to_cut_and_build_matix = generate_words_matrix(not_use40)
```

```python
f_indoc = {}

for i in use_this_dict_to_cut_and_build_matix:
    f_indoc[i] = 0
```

```python
for idx, row in df.iterrows():
    text = row['Text']
    words = []
    for w in use_this_dict_to_cut_and_build_matix:
        if w in text:
            words.append(w)

    set_of_words = set(words)
    for w in set_of_words:
        if w in f_indoc:
            f_indoc[w] += 1
        else:
            continue
```

```python
import math

words_idf = {}

for w in f_indoc:
    words_idf[w] = math.log10(130/f_indoc[w])
```

```python
with open('./final_process/words_idf.pickle', 'wb') as f:
    pickle.dump(words_idf, f, pickle.HIGHEST_PROTOCOL)
```

```python
tfidf_per_NO = {}

for idx, row in df.iterrows():
    w_indoc = {}
    ll = 0
    text = row['Text']
    for w in use_this_dict_to_cut_and_build_matix:
        if w in row['Text']:
            if w in w_indoc:
                w_indoc[w] += text.count(w)
            else:
                w_indoc[w] = text.count(w)
            ll += text.count(w)
            
    for w in w_indoc:
        w_indoc[w] = w_indoc[w]/ll
        
    xx = {}
    for w in words_idf:
        if w in w_indoc:
            xx[w] = words_idf[w] * w_indoc[w]
        else:
            xx[w] = 0
    tfidf_per_NO[row['NO']] = xx
```

```python
with open('./final_process/tf_idf_no.pickle', 'wb') as f:
    pickle.dump(tfidf_per_NO, f, pickle.HIGHEST_PROTOCOL)
```
