```python
import pandas as pd
from pythainlp.tokenize import word_tokenize
```

```python
df = pd.read_csv('./final_process/law_text.csv')
df.head()
```

```python
# import deepcut

wc = []
ts = []
tns = []

for idx, row in df.iterrows():
#     words = deepcut.tokenize(row['Text'])
    words = word_tokenize(row['Text'])
    wc.append(len(words))
    ts.append(' '.join(words))
    temp = ''
    for i in words:
        if i != ' ':
            temp += i
    tns.append(temp)
```

```python
df['word_count'] = wc
df['text_no_space'] = tns
df['text_space_pythai'] = ts
```

```python
## use sorted_dict

import pickle

file = open('./poc/relate_word_dict.pickle','rb')
relate_dict = pickle.load(file)
file.close()

def generate_words_matrix(not_use_arr):
    words_for_matrix = []
    for i in sorted_dict:
        if i not in not_use_arr:
            words_for_matrix.append(i)
    temp = []
    for i in relate_dict:
        if i in words_for_matrix:
            for j in relate_dict[i]:
                temp.append(j)
    words_for_matrix += temp
    words_for_matrix = set(words_for_matrix)
    words_for_matrix = list(words_for_matrix)
    return words_for_matrix
```

```python
f_indoc = {}

for i in use_this_dict_to_cut_and_build_matix:
    f_indoc[i] = 0
```

```python
for idx, row in df.iterrows():
    text = row['Text']
    words = []
    for w in use_this_dict_to_cut_and_build_matix:
        if w in text:
            words.append(w)

    set_of_words = set(words)
    for w in set_of_words:
        if w in f_indoc and w in sorted_by_law:
            f_indoc[w] += 1
        else:
            continue
```

```python
import math

words_idf = {}

for w in f_indoc:
    try:
        words_idf[w] = math.log10(130/f_indoc[w])
    except:
        words_idf[w] = 0
```

```python
tfidf_per_NO = {}

for idx, row in df.iterrows():
    w_indoc = {}
    ll = 0
    text = row['Text']
    for w in use_this_dict_to_cut_and_build_matix:
        if w in row['Text']:
            if w in w_indoc:
                w_indoc[w] += text.count(w)
            else:
                w_indoc[w] = text.count(w)
            ll += text.count(w)
            
    for w in w_indoc:
        w_indoc[w] = w_indoc[w]/ll
        
    xx = {}
    for w in words_idf:
        if w in w_indoc:
            xx[w] = words_idf[w] * w_indoc[w]
        else:
            xx[w] = 0
    tfidf_per_NO[row['NO']] = xx
```

```python
## process relate words
## ต้องไม่อยู่ในคำหลัก แล้วก็ต้องเท่ากับ 0 ไป *0.8
## ถ้าไม่เท่ากับ 0 ควรจะเอาค่ามันไป *0.8 ด้วยป่ะ เพราะมันไม่ใช่คำที่มีข้างใน

for i in tfidf_per_NO:
    a = tfidf_per_NO[i]
#     print(a)
    for j in relate_dict:
        if j in a and a[j] > 0:
            for k in relate_dict[j]:
                if k not in sorted_by_law and k in use_this_dict_to_cut_and_build_matix and a[k] == 0:
                    a[k] = a[j] * 0.8
```
