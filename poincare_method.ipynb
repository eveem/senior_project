```python
import pandas as pd

df = pd.read_csv('police_record.csv')
```

```python
from pythainlp.tokenize import word_tokenize

wdf = pd.read_csv('project_words_edit.csv')
```

```python
data = {}

for index, row in wdf.iterrows():
    if type(row['Verb']) is str:
        a = set()
        x = row['Verb'].split(', ')
        for i in x:
            jj = word_tokenize(i)
            for j in jj:
                a.add(j)
    
    if type(row['Noun']) is str:
        b = set()
        x = row['Noun'].split(', ')
        for i in x:
            jj = word_tokenize(i)
            for j in jj:
                b.add(j)
    
    if type(row['Extra']) is str:
        c = set()
        x = row['Extra'].split(', ')
        for i in x:
            jj = word_tokenize(i)
            for j in jj:
                c.add(j)
    if type(row['Penal No.']) is str or type(row['Penal No.']) is int:
        data[str(row['Penal No.'])] = [a, b, c]
```

```python
for index, row in wdf.iterrows():
    if type(row['Inherit [เอา v + n ข้อนั้น แต่ต้อง ^ extra]']) is str:
        x = row['Inherit [เอา v + n ข้อนั้น แต่ต้อง ^ extra]'].split(', ')
        a = []
        b = []
        c = []
        for i in x:
            if i in data:
                a += data[i][0]
                b += data[i][1]
                c += data[i][2]
        data[row['Penal No.']] = [a, b, c]
```

```python
## poincare + wordnet
 
from pythainlp.corpus import wordnet
import pickle

model = pickle.load(open('./poc/v_n_poincare.pickle', 'rb'))
nd = {}

for i in data:
    g = []
    for j in data[i]:
        for k in j:
            try:
                g += model.kv.descendants(k)
            except:
                g.append(k)
    nd[i] = g

nnd = {}

for i in nd:
    g = []
    for j in nd[i]:
        try:
            g += wordnet.synsets(j)[0].lemma_names('tha')
        except:
            g.append(j)
    nnd[i] = g
```

```python
import operator

point = {}

for index, row in df.iterrows():
    k = []
    r = []
    text = row['Conclusion']
    words = word_tokenize(text)
    
    for d in nnd:
        ck = 0
        for w in words:
            if w in nnd[d]:
                ck += 1
        k.append(d)
        r.append(ck)
    rdata = {'Code': k, 'Counter': r} 
    rdf = pd.DataFrame(rdata)
    rdf = rdf.sort_values(by=['Counter'], ascending=False)
```

```python
dd = []

for index, row in df.iterrows():
    x = row['Labels']
    x = x.replace('[', '')
    x = x.replace(']', '')
    x = x.replace('\'', '')
    x = x.split(', ')
    dd.append(x)
```
