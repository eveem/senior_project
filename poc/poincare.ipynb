```python
from gensim.models.poincare import PoincareModel
import pickle
```

```python
v_relation = pickle.load(open('verbpaire.pickle', 'rb'))
```

```python
v_relation
```

```python
v_model = PoincareModel(v_relation, negative=2)
```

```python
v_model.train(epochs=100)
```

# Verb only

```python
v_model.kv.descendants('บังคับ')
```

```python
v_model.kv.descendants('โกง')
```

# Noun only

```python
n_relation = pickle.load(open('nounpair.pickle', 'rb'))
```

```python
n_model = PoincareModel(n_relation, negative=2)
```

```python
n_model.train(epochs=100)
```

```python
n_model.kv.descendants('เจ้าหน้าที่')
```

```python
n_model.kv.descendants('อาวุธ')
```

# Verb + Noun

```python
relations = n_relation + v_relation
```

```python
mix_model = PoincareModel(relations, negative=2)
```

```python
mix_model.train(epochs=100)
```

```python
mix_model.kv.descendants('บังคับ')
```

```python
mix_model.kv.descendants('โกง')
```
