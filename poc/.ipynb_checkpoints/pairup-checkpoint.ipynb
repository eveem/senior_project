```python
import pandas as pd
import re

df = pd.read_csv('./verbunq.csv')
```

```python
x = []

for index, row in df.iterrows():
    for i in row['related word'].split(', '):
        j = (row['word'], i)
        x.append(j)
```

```python
import pickle

pickle.dump(x, open('verbpaire.pickle', 'wb'))
```
